### Step 2: Download Swiss-Prot Database

This cell downloads and decompresses the complete Swiss-Prot database.

```
import requests
import gzip
import os

url = "https://ftp.uniprot.org/pub/databases/uniprot/current_release/knowledgebase/complete/uniprot_sprot.fasta.gz"
output_gz_file = "uniprot_sprot.fasta.gz"
output_fasta_file = "uniprot_sprot.fasta"

print(f"Downloading Swiss-Prot database...")
response = requests.get(url, stream=True)
response.raise_for_status()
with open(output_gz_file, 'wb') as f:
    for chunk in response.iter_content(chunk_size=8192):
        f.write(chunk)

print(f"Decompressing...")
with gzip.open(output_gz_file, 'rb') as f_in:
    with open(output_fasta_file, 'wb') as f_out:
        f_out.write(f_in.read())
os.remove(output_gz_file)
print(f"✅ {output_fasta_file} is ready.")
```

### Step 4: Compute Database Embeddings (The Long Step)

This cell will take **many hours** to run as it processes the entire Swiss-Prot database. It will save its progress as it goes.

```
import torch
import esm
from Bio import SeqIO
import time

FASTA_FILE = "uniprot_sprot.fasta"
EMBEDDINGS_OUTPUT_FILE = "swissprot_embeddings.pt"
CHECKPOINT_INTERVAL = 5000

print("Loading ESM-2 model...")
model, alphabet = esm.pretrained.esm2_t33_650M_UR50D()
model.eval()
model = model.cuda()
batch_converter = alphabet.get_batch_converter()
print("Model loaded to GPU.")

if os.path.exists(EMBEDDINGS_OUTPUT_FILE):
    embeddings = torch.load(EMBEDDINGS_OUTPUT_FILE)
    processed_ids = set(embeddings.keys())
else:
    embeddings = {}
    processed_ids = set()
print(f"Starting/Resuming embedding generation. {len(processed_ids)} sequences already processed.")

start_time = time.time()
sequences_processed_this_run = 0
for i, seq_record in enumerate(SeqIO.parse(FASTA_FILE, "fasta")):
    if seq_record.id in processed_ids:
        continue
    
    data = [(seq_record.id, str(seq_record.seq))]
    batch_labels, batch_strs, batch_tokens = batch_converter(data)
    batch_tokens = batch_tokens.cuda()
    with torch.no_grad():
        results = model(batch_tokens, repr_layers=[33])
    token_representations = results["representations"][33]
    sequence_representation = token_representations[0, 1 : len(str(seq_record.seq)) + 1].mean(0).to("cpu")
    embeddings[seq_record.id] = sequence_representation
    sequences_processed_this_run += 1

    if sequences_processed_this_run > 0 and sequences_processed_this_run % CHECKPOINT_INTERVAL == 0:
        elapsed = time.time() - start_time
        print(f"  - Processed {sequences_processed_this_run} new sequences in {elapsed:.2f}s. Total: {len(embeddings)}. Saving checkpoint...")
        torch.save(embeddings, EMBEDDINGS_OUTPUT_FILE)
        start_time = time.time()

print("\n✅ Embedding generation complete.")
torch.save(embeddings, EMBEDDINGS_OUTPUT_FILE)
print("Final embeddings file saved.")
```

### Step 5: Build Search Index

This cell builds the fast search index from the database embeddings.

```
import faiss
import numpy as np

INDEX_FILE = "swissprot_faiss.index"
ID_MAP_FILE = "swissprot_id_map.pt"

print(f"Loading embeddings from swissprot_embeddings.pt...")
embeddings_data = torch.load("swissprot_embeddings.pt")

id_map = list(embeddings_data.keys())
embedding_matrix = np.array([embeddings_data[id].numpy() for id in id_map], dtype='float32')
dimension = embedding_matrix.shape[1]

print(f"Building FAISS index...")
index = faiss.IndexFlatL2(dimension)
# If you have a GPU and faiss-gpu is installed, you can use:
    # res = faiss.StandardGpuResources()
    # index = faiss.index_cpu_to_gpu(res, 0, index)
    index.add(embedding_matrix)

print(f"Saving index and ID map...")
faiss.write_index(index, INDEX_FILE)
torch.save(id_map, ID_MAP_FILE)
print("✅ Indexing complete.")
```

### Step 6: Find and Display Top Hits

This final cell performs the search and prints your results.

```
import pandas as pd

print("Loading all files for final search...")
index = faiss.read_index("swissprot_faiss.index")
id_map = torch.load("swissprot_id_map.pt")
orphan_embeddings_data = torch.load("embeddings.pt")
target_df = pd.read_csv("target20.csv", sep='\t', header=None, names=['Receptor', 'PDB_ID', 'DOI', 'Year', 'Key_Mystery'])
receptor_names = {f"{row['PDB_ID']}_A": row['Receptor'] for index, row in target_df.iterrows()}

results = {}
for orphan_key, orphan_embedding in orphan_embeddings_data.items():
    if orphan_key not in receptor_names:
        continue
    
    query_vector = orphan_embedding.numpy().reshape(1, -1).astype('float32')
    distances, indices = index.search(query_vector, 5)
    top_hits = [id_map[i] for i in indices[0]]
    results[receptor_names[orphan_key]] = top_hits

print("\n--- FINAL RESULTS: AI-Predicted Functional Relatives ---")
print("{:<15} | {:<25} | {:<70}".format("Orphan Receptor", "Top Predicted Relative", "UniProt URL"))
print("-" * 115)

for receptor, hits in sorted(results.items()):
    top_hit_id = hits[0]
    accession_id = top_hit_id.split('|')[1]
    url = f"https://www.uniprot.org/uniprot/{accession_id}"
    print("{:<15} | {:<25} | {:<70}".format(receptor, accession_id, url))
```
